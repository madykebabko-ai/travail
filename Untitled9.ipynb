```python
from google.colab import files
import json

uploaded = files.upload()

# Lire le fichier uploadé (peu importe son nom)
filename = list(uploaded.keys())[0]
donnees = json.loads(uploaded[filename].decode('utf-8'))

print("✅ Données reçues :")
for d in donnees:
    print(f"   {d}")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "Qwen/Qwen2.5-1.5B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16, device_map="auto")
print("✅ Qwen chargé !")
```

```python
import json

resultats = []
for d in donnees:
    risque = "Élevé" if d['montant'] > 20000 else "Faible"
    analyse = "Transaction à surveiller" if risque == "Élevé" else "Transaction normale"
    resultats.append({
        "id": d["id"],
        "nomclient": d["nomclient"],
        "date": d["date"],
        "montant": d["montant"],
        "analyse": analyse,
        "risque": risque
    })
    print(f"✅ Client {d['id']} analysé - Risque : {risque}")
```

```python
from google.colab import files

with open("resultat_colab.json", "w", encoding="utf-8") as f:
    json.dump(resultats, f, ensure_ascii=False, indent=2)

files.download("resultat_colab.json")
print("✅ resultat_colab.json téléchargé !")
```
